```python

resume_text = data.lower()
resume_text = re.sub(r'[^a-z0-9\s]', ' ', resume_text)
resume_text = re.sub(r'\s+', ' ', resume_text)
resume_text = resume_text.replace("\n", " ")
```

```python
matched_keywords = []
missing_keywords = []

for keyword in jd_keywords:
    if keyword in resume_text:
        matched_keywords.append(keyword)
    else:
        missing_keywords.append(keyword)
```

```python
total_keywords = len(jd_keywords)
matched_count = len(matched_keywords)

resume_score = round((matched_count / total_keywords) * 100, 2)

print("Resume Score:", resume_score, "%")
```

```python
for i in range(n):
    word = dic['text'][i].strip().lower()

    if word == "":
        continue

   
    word = re.sub(r'[^a-z0-9]', '', word)

    
    for key in jd_keywords:
        if word in key.split():
            print(dic['text'][i])

            x = dic['left'][i]
            y = dic['top'][i]
            w = dic['width'][i]
            h = dic['height'][i]

            cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 3)
            break
```
